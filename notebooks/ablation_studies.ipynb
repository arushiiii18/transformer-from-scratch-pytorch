```python
import json
import matplotlib.pyplot as plt
import numpy as np

with open('../ablation_results/results.json') as f:
    results = json.load(f)

def get_final_val(name):
    return results[name]['results'][-1]['val_loss']

fig, axes = plt.subplots(1, 3, figsize=(18, 5))
fig.suptitle('Ablation Studies — Effect of Architectural Choices on Val Loss', fontsize=13)

heads = [1, 2, 4, 8]
head_losses = [get_final_val(f'heads_{h}') for h in heads]
axes[0].plot(heads, head_losses, 'bo-', linewidth=2, markersize=8)
axes[0].set_xlabel('Number of Attention Heads')
axes[0].set_ylabel('Validation Loss')
axes[0].set_title('Attention Heads vs Val Loss')
axes[0].set_xticks(heads)
axes[0].grid(True, alpha=0.3)
for h, l in zip(heads, head_losses):
    axes[0].annotate(f'{l:.3f}', (h, l), textcoords='offset points', xytext=(0,10), ha='center')

layers = [1, 2, 3, 6]
layer_losses = [get_final_val(f'layers_{n}') for n in layers]
axes[1].plot(layers, layer_losses, 'ro-', linewidth=2, markersize=8)
axes[1].set_xlabel('Number of Encoder/Decoder Layers (N)')
axes[1].set_ylabel('Validation Loss')
axes[1].set_title('Depth vs Val Loss')
axes[1].set_xticks(layers)
axes[1].grid(True, alpha=0.3)
for n, l in zip(layers, layer_losses):
    axes[1].annotate(f'{l:.3f}', (n, l), textcoords='offset points', xytext=(0,10), ha='center')

dffs = [256, 512, 1024, 2048]
dff_losses = [get_final_val(f'dff_{d}') for d in dffs]
axes[2].plot(dffs, dff_losses, 'go-', linewidth=2, markersize=8)
axes[2].set_xlabel('FFN Hidden Dimension (d_ff)')
axes[2].set_ylabel('Validation Loss')
axes[2].set_title('FFN Size vs Val Loss')
axes[2].set_xticks(dffs)
axes[2].grid(True, alpha=0.3)
for d, l in zip(dffs, dff_losses):
    axes[2].annotate(f'{l:.3f}', (d, l), textcoords='offset points', xytext=(0,10), ha='center')

plt.tight_layout()
plt.savefig('../ablation_plots.png', dpi=150, bbox_inches='tight')
plt.show()
print('Saved to ablation_plots.png')
```
